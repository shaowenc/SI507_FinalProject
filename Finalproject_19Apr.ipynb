### 2nd data source: applying OpenWeather API to collect weather data for each cities and output to a weather.csv

```python
cities = []
country = []
lat = []
lon = []
temperature = []
humidity = []
cloudiness = []
wind_speed = []
date = []
```

```python
cities_pd = pd.read_csv("./worldcities.csv")
for i in range(len(cities_pd)):
    city_name = cities_pd.iloc[i]["city_ascii"]
    country_code = cities_pd.iloc[i]["country"]
    
    response = requests.get(f"http://api.openweathermap.org/data/2.5/weather?q={city_name},{country_code}&appid=cc1fc5169c434d0a5df140385c2df47c").json()
    #print(i, response['name'])
    cities.append(response['name'])
    country.append(response['sys']['country'])
    lat.append(response['coord']['lat'])
    lon.append(response['coord']['lon'])
    temperature.append(response['main']['temp_max'])
    humidity.append(response['main']['humidity'])
    cloudiness.append(response['clouds']["all"])
    wind_speed.append(response['wind']['speed'])
    date.append(response['dt'])
```

```python
weather_dict = {
    "City": cities,
    "Country": country,
    "Latitude": lat,
    "Longitude": lon,
    "Max Temp (F)": temperature,
    "Humidity %": humidity,
    "Cloudiness %": cloudiness,
    "Wind Speed (mph)": wind_speed,
    "Date": date
}
weather_data = pd.DataFrame(weather_dict)
weather_data

# Export weather as a CSV
weather_data.to_csv("./Weather_Data.csv", index=False, header=True)
```
